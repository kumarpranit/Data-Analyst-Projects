# Assignment Number 6.6 Data Science: Duplicate Word Removal
**Exercise Description** (Duplicate Word Removal) 
- Write a function that receives a list of words, then determines and returns a sorted list of the unique words.
- Treat uppercase and lowercase the same—for example, if the list contains *apple* and *Apple*, the list you return should contain *apple* once.
- The caller should display the unique words in alphabetical order with one word per line. Test your function by calling it several times with different sentences. Remember to break each sentence into a list of words before passing it to the function.

**LabAssist:** 
- The techniques used here are data-wrangling operations.


```python
def return_sorted_unique_list(sentence):
    """ This function returns a sorted list of the unique words from a given sentence."""
    words_lower = [words.lower() for words in sentence]
    unique_words = set(words_lower) ##unique words in the given list
    return(sorted(unique_words))
```

```python
input_sentence = 'My favorite subject subject Is Is Python'
print(f'The input sentence is: {input_sentence}')
print('------ Output below----------')
##Split the sentence into individual words and pass it as a list
split_sentence = input_sentence.split()
sorted_unique  = return_sorted_unique_list(split_sentence)
for words in sorted_unique:
    print(words)
```

```python
input_sentence = 'Dodgers GAME game was full FULL OF ENTERTAINMENT entertainment'
print(f'The input sentence is: {input_sentence}')
print('------ Output below----------')

##Split the sentence into individual words and pass it as a list
split_sentence = input_sentence.split()
sorted_unique  = return_sorted_unique_list(split_sentence)
for words in sorted_unique:
    print(words)
```

- Converts all the words in the sentence to lowercase
- Extracts a unique set of words i.e. it removes the duplicates word in the sentence
       - The **set** function hlps to removes the duplicate

**Main Code**
- The input sentence whose words have to be sorted and printed is passed to the defined function **return_sorted_unique_list**
- The list obtained is then iterated to print out the words in lexicographical order
