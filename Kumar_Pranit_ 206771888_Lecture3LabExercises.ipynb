# Assignment Number 4.9 Data Science: Temperature Conversion

## Problem Statement:
Implement a `fahrenheit` function that returns the Fahrenheit equivalent of a Celsius temperature.  

The formula to use is:  

F = (9 / 5) * C + 32

## LabAssist notes:
1. Implement the `fahrenheit` function.
2. Use this function to print a chart showing the Fahrenheit equivalents of all Celsius temperatures in the range **0–100 degrees**.
3. Use **one digit of precision** for the results.
4. Print the outputs in a **neat tabular format**.


```python
from decimal import Decimal
```

```python
## Logic for converting celcius to fahrenheit using a function
def temperature_conversion(celsius):
    fah = Decimal((9 / 5)) * (Decimal(celsius)) + 32
    return fah
```

```python
##Display headers for the table
print(f"{' Celsius ':-^11}|{' Fahrenheit ':-^14}")
print("-" * 26) 

##Implementation part to display the data in tabular format
for celsius in range(1,101):
    fahrein = temperature_conversion(celsius) ##Function call to convert each celsius degree to corresponding fahrenheit
    print(f"{celsius:^11}|{fahrein:>9.1f} ") 
```

# Assignment Number 5.7 Duplicate Elimination
- This example filters the unique values present in a given list:  
## LabAssist Notes:
- For each item in the original list, use not in to determine whether the item is in the new list
- If not, add it. Do not use **NumPy’s unique function** in your solution

```python
def return_unique_list(new_list):
    new_unique = []
    for items in new_list:
        if items not in new_unique: ##check if the items in the list is present or not in the new one
            new_unique.append(items) ##append the new list if the item is not present

    new_unique.sort()
    return list(new_unique)
```

# Assignment Number 5.14 (Is a Sequence Sorted?) 
- Create a function is_ordered that receives a sequence and returns True if the elements are in sorted order.
- Test your function with sorted and unsorted lists, tuples and strings.

## LabAssist Notes:
- Do not use the below logic for sorting:
  - Create a sorted copy of the sequence using built-in function sorted, then compare the copy to the original
  - Sorting is an “expensive” operation, especially as the number of elements to be sorted increases

```python
## Function to check if the sequence provided is sorted or not?

def is_sequence_sorted(sequence):
    check_sorted = ''
    items = 0
    while items != len(sequence)-1:
        if sequence[items] <= sequence[items+1]:
            items += 1
            check_sorted = 'True'
            continue
        else:
            check_sorted = 'False'
            return check_sorted
    return check_sorted
```

```python
list_of_numbers = [1,1,1,1,1,6,3,6,6,4]
check_sort = is_sequence_sorted(list_of_numbers) ##function call 
if check_sort == 'True':
    print(f'The provided list {list_of_numbers}  is in a sorted sequence')
else:
    print(f'The provided list {list_of_numbers}  is not a sorted sequence')
```

```python
list_string = ['a','A','c','d','e','f']
check_sort = is_sequence_sorted(list_string) ##function call 
if check_sort == 'True':
    print(f'The provided list {list_string}  is in a sorted sequence')
else:
    print(f'The provided list {list_string}  is not a sorted sequence')
```
