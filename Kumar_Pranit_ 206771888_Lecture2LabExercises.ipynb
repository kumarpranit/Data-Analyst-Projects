# Assignment Number 3.10(7% Investment Return)

## The original assignment
Some investment advisors say that it’s reasonable to expect a **7% return** over the long term in the stock market.  

Assuming an initial investment of **$1000**, we calculate the amount of money available after **10, 20, and 30 years** using the formula: 

\[
a = p * (1 + r) ** n
\]  

Where:  
- `p` → the principal (original amount invested)  
- `r` → the annual rate of return (7% = 0.07)  
- `n` → the number of years (10, 20, 30) 
- `a` → the amount on deposit at the end of `n` years

## Lab Assist Notes/ Instructions
1. Reimplement this exercise to use a loop that calculates and displays the amount of money you’ll have each year at the end of years 1 through 30

## Developer Comments
* The displayed amount is restricted to three decimal places.
* The f-string (short for formatted string) allows for the insertion of values into a string. It is denoted by an "f" before the opening quote.
* The curly braces in the print statement specify where to insert the values, using `{}`.

````python
##Initialisation of variables##

principal_amount = 1000
rate_return_annual = 0.07
````

````python
##Processing and Execution##

for years in range ( 1, 31 ):
    amount_at_end = principal_amount * ( 1 + rate_return_annual ) ** years
    if years == 1:
        ##Reformatting the print statement for the first year only
        print(f'Amount Invested after {years} year is: $ {amount_at_end:.3f}')
    else:
        print(f'Amount Invested after {years} years is: $ {amount_at_end:.3f}')
````

## Measures of Central Tendency for Categorical Data

When working with **categorical (non-numeric)** data, the most appropriate measure of central tendency is the **Mode**. Unlike numerical data, categorical values cannot be averaged or ranked meaningfully. The **Mode** identifies the category that occurs most frequently, offering valuable insights into dominant preferences, trends, or classifications within the dataset.

---

## Example: Favorite Motorcycle Brands

Let's examine a dataset capturing individuals' favorite motorcycle brands:

| Person | Favorite Motorcycle Brand |
|--------|---------------------------|
| 1      | Harley-Davidson           |
| 2      | Honda                     |
| 3      | Kawasaki                  |
| 4      | Harley-Davidson           |
| 5      | Harley-Davidson           |

In this example, the **mode** is **Harley-Davidson**, as it appears **three times**, more than any other brand. This indicates that Harley-Davidson is the most preferred motorcycle brand among the group surveyed.

---

✅ **Key Takeaway**:  
For categorical data, identifying the mode helps highlight the most common category, which can inform decisions, marketing strategies, or further analysis.
