# Circle Calculations: Area, Diameter, and Circumference - Assignment Number 2.5

This assignment demonstrates how to calculate the **diameter**, **circumference**, and **area** of a circle given a radius.  

For this assignment, we‚Äôll use a **circle of radius `2`**, and approximate œÄ as **3.14159**.

---

## üìê Formulas used for calculation

- **Diameter** = 2 * r

- **Circumference** = 2 * pi * r 

- **Area** =  pi * r ** 2

Where:
- \( r \) = radius of the circle  
- \( pi ) ‚âà 3.14159  


```python
radius_of_circle = 2
```

```python
diameter_of_circle = 2 * radius_of_circle
```

```python
print("Diameter of Circle:", diameter_of_circle)
```

```python
value_of_pi = 3.14159
```

```python
circumference_of_circle = 2 * value_of_pi * radius_of_circle
```

```python
print("Circumference of Circle:", circumference_of_circle)
```

```python
area_of_circle = value_of_pi * radius_of_circle ** 2
```

```python
print("Area of Circle:", area_of_circle)
```

# üí∞7% Investment Return ‚Äî Assignment Number 2.12  

Some investment advisors say that it‚Äôs reasonable to expect a **7% return** over the long term in the stock market.  

Assuming an initial investment of **$1000**, we calculate the amount of money available after **10, 20, and 30 years** using the formula:  

\[
a = p * (1 + r) ** n
\]  

Where:  
- `p` ‚Üí the principal (original amount invested)  
- `r` ‚Üí the annual rate of return (7% = 0.07)  
- `n` ‚Üí the number of years (10, 20, 30)  
- `a` ‚Üí the amount on deposit at the end of `n` years

## üìå Instructions  
1. Prompt the user to input number of investment duration.
2. The value of investment duration should only be 10,20 or 30  

```python
principal_amount = 1000
```

```python
rate_return_annual = 0.07
```

```python
number_of_years = int(input("Please enter the investment duration in years: "))
```

```python
amount_at_end = principal_amount * ( 1 + rate_return_annual ) ** number_of_years
```

```python
print("Investment after", number_of_years, "years is: $", amount_at_end)
```
